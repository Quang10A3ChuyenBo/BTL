```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt

# Đọc dữ liệu từ file CSV
commits_df = pd.read_csv('commits.csv')

# Tạo đồ thị
G = nx.Graph()

# Thêm các nodes (contributors) vào đồ thị
for contributor in commits_df['Contributor'].unique():
    G.add_node(contributor)

# Thêm các cạnh (edges) giữa các contributors dựa trên sự cộng tác
for _, row in commits_df.iterrows():
    G.add_edge(row['Author'], row['Contributor'])

# Tính toán độ trung tâm về bậc
degree_centrality = nx.degree_centrality(G)

# Vẽ đồ thị
plt.figure(figsize=(12, 8))
pos = nx.spring_layout(G, k=0.5, iterations=100)

# Vẽ các node với kích thước tương ứng với độ trung tâm về bậc
node_sizes = [degree_centrality[node] * 5000 for node in G.nodes()]
nx.draw(G, pos, with_labels=True, node_color='lightblue', edge_color='gray', node_size=node_sizes, font_size=10)

plt.title('Mạng lưới cộng tác giữa các contributors')
plt.show()

# Tìm người có nhiều kết nối nhất
most_connected = max(degree_centrality, key=degree_centrality.get)
print(f"Người có nhiều kết nối nhất: {most_connected}")
```
